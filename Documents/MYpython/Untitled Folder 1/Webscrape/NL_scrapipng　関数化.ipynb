```python
def NLget(urladdress):
    
    url=urladdress
    res=requests.get(url)
    soup = BeautifulSoup(res.text, "html.parser")
    res.text
    markup = res.text#OKエラーがでるので調整済み
    soup = BeautifulSoup(markup, 'lxml')

    nl_text=soup.find_all('article')
    partzero=nl_text[0].text
    print(partzero)#ＯＫ 記事とれた 日本語もOK
    with open('animals.txt', mode='w', encoding='utf-8') as fout:#textファイル書き込み
        fout.write(partzero)
        fout.flush()
```

```python
NLget('https://www.label-bank.co.jp/blog/foodlabel/202012nutrition')#関数呼び出し
```
