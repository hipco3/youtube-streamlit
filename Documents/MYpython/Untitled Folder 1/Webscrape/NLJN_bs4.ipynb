```python
!pip install requests
!pip install lxml
!pip install BeautifulSoup4
import requests
from bs4 import BeautifulSoup
from html.parser import HTMLParser


url='https://www.label-bank.co.jp/blog/foodlabel/202102foshu'
res=requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")
res.text
markup = res.text#OKエラーがでるので調整済み
soup = BeautifulSoup(markup, 'lxml')

nl_text=soup.find_all('article')
partzero=nl_text[0].text
print(partzero)#ＯＫ 記事とれた 日本語もOK
with open('animals.txt', mode='w', encoding='utf-8') as fout:#textファイル書き込み
    fout.write(partzero)
    fout.flush()
```

```python
#違う要素でとった結果
import pandas as pd
partone = soup.find_all('div', attrs={'id':'content'})
partone=partone[0].text
print(partone)
with open('animals.xlsx', mode='w', encoding='utf-8') as fout:#textファイル書き込み
    fout.write(partone)
    fout.flush()
```
